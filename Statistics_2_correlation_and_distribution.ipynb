```python
price = corr['MEDV']
room = corr['RM']
```

```python
price = np.array(price).reshape(-1,1)
room = np.array(room).reshape(-1,1)
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
```

```python
#тренируем
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    room,
    price,
    test_size = 0.2
)
```

```python
linreg.fit(X_train, y_train)
```

```python
predicted_y = linreg.predict(X_test)
linreg.score(X_test, y_test)
```

```python
plt.xlabel('Количество комнат')
plt.ylabel('Цена')
plt.title('Линейная регрессия от одной переменной')
x = np.linspace(0, 10, 100)
# k = model.coef_
# b = model.intercept_
y = linreg.coef_*x + linreg.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(1, 10)
plt.ylim(0, 60)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
number_of_randoms = 100
number_of_sums = 1000
X1 = np.zeros([number_of_sums])
for i in range(number_of_sums):
    X1[i] = np.sum(np.random.rand(1,6,number_of_randoms))
plt.hist(X1, 100)
plt.xlabel('Частота суммы')
plt.ylabel('Сумма бросков')
plt.show()
# Нам было необходимо достаточное кол-во данных для нормального распределение и применения ЦПТ
```
