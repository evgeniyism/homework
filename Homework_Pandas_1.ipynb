# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
rate = pd.read_csv('ratings.csv')
mov =  pd.read_csv('movies.csv')
```

```python
full = rate.merge(mov, on ='movieId', how = 'left')
full = full.drop(['genres','timestamp',], axis='columns') 
full = full[full['rating'] == 5]
full['rating'].astype('float')
full['RatingCount'] = full.groupby('movieId').sum()[['rating']] # какая-то черная магия, 
                                                                # не понимаю до конца логику этой строки
full = full.sort_values(by = 'RatingCount', ascending = False)
full.head(1)
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
```

```python
power['category'].astype('int')
power = power[( (power['category'] == 4) | (power['category'] == 12) | (power['category'] == 21) ) 
              & ( (power['year'] >= 2005) & (power['year'] <= 2010) ) &
             ( (power['country'] == 'Lithuania') | (power['country'] == 'Latvia') | (power['country'] == 'Estonia') )]

result = power.groupby('country').sum()[['quantity']]
result
```
